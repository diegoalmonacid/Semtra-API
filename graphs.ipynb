```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargar los datos del archivo JSON
with open("test-10-5min-output.json") as file:
    data = json.load(file)
# Extraer tiempos de respuesta medios de cada intervalo
intervals = data["intermediate"]
response_times = [interval["summaries"]["http.response_time.2xx"]["mean"] for interval in intervals]
timestamps = range(len(response_times))  # Cada intervalo representa un periodo

# Grfico de tiempos de respuesta
plt.figure(figsize=(10, 5))
plt.plot(timestamps, response_times, marker='o', label="Tiempo medio de respuesta (ms)")
plt.title("Evolucin de los Tiempos de Respuesta")
plt.xlabel("Intervalo de Tiempo (10 segundos por intervalo)")
plt.ylabel("Tiempo Medio de Respuesta (ms)")
plt.grid(True)
plt.legend()
plt.show()
```

```python


# Extraer todos los tiempos mnimos, mximos y medios
response_min = [interval["summaries"]["http.response_time"]["min"] for interval in intervals]
response_max = [interval["summaries"]["http.response_time"]["max"] for interval in intervals]

# Crear un histograma
plt.figure(figsize=(10, 5))
plt.hist(response_min + response_max + response_times, bins=10, color='blue', alpha=0.7, label="Distribucin de tiempos")
plt.title("Distribucin de los Tiempos de Respuesta")
plt.xlabel("Tiempo de Respuesta (ms)")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.show()
```
